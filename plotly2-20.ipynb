```python
pokedata = pd.DataFrame(pokedex['pokemon'])
pokedata.head()

#DataFrame is Case Sensitive!!! Dumdum.
```

```python
pokedata[['type1','type2']] = pd.DataFrame(pokedata.type.tolist(), index= pokedata.index)
pokedata.head(10)
 
```

```python
typegroup = pokedata.groupby('type1').count()['id']
 
typegroup
```

```python
fig = go.Figure(data=[go.Bar(x=typegroup.index, y=typegroup.values)])
fig.show()
```
