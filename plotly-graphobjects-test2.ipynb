```python
#now that we know there's only one key in a list of dictionaries, let's make a dataframe
pokedata = pd.DataFrame(pokedex['pokemon'])
#case sensitive
pokedata.head(10)
```

```python
pokedata[['type1','type2']] = pd.DataFrame(pokedata.type.tolist(), index= pokedata.index)
pokedata.head(10)
```

```python
typegroup = pokedata.groupby('type1').count()['id']
#now we have types categorized we can make our first bar graph!
typegroup
```

```python
fig1 = go.Figure(data=[go.Bar(x=typegroup.index, y=typegroup.values)])

fig1.show()
 
```

```python
pokecolors_map = {
	'Normal': '#A8A77A',
	'Fire': '#EE8130',
	'Water': '#6390F0',
	'Electric': '#F7D02C',
	'Grass': '#7AC74C',
	'Ice': '#96D9D6',
	'Fighting': '#C22E28',
	'Poison': '#A33EA1',
	'Ground': '#E2BF65',
	'Flying': '#A98FF3',
	'Psychic': '#F95587',
	'Bug': '#A6B91A',
	'Rock': '#B6A136',
	'Ghost': '#735797',
	'Dragon': '#6F35FC',
	'Dark': '#705746',
	'Steel': '#B7B7CE',
	'Fairy': '#D685AD',
};

pokecolors = typegroup.index.map(pokecolors_map)
pokecolors= list(pokecolors)
pokecolors
```

```python
type_names = list(typegroup.index)
type(type_names)
type_names
```

```python
fig1 = go.Figure(data=[go.Bar(x=type_names, y=typegroup.values,
 marker_color=pokecolors)])
#how cool! but we can make this graph a bit more helpful by making it a pie chart
fig1.show()
```

```python
fig = go.Figure(
    data=[
    go.Pie(
        labels=typegroup.index, 
        values=typegroup.values,
        marker=dict(colors=pokecolors),
        title="Gen 1 Pokemon by Type"
        )])


fig.show()
```

```python
pokedata.tail()
#cool so our next step to tidy is sorting evolution stages
```

```python
type(pokedata['next_evolution'][150])
```

```python
def get_stage(row):
    if isinstance(row['next_evolution'], list) and isinstance(row['prev_evolution'], float):
        return "Starter"
    elif isinstance(row['next_evolution'], float) and isinstance(row['prev_evolution'], list):
        return "Final"
    elif isinstance(row['next_evolution'], float) and isinstance(row['prev_evolution'], float):
        return "Lone"
    else:
        return "Middle"

pokedata['evolution_stage'] = pokedata.apply(get_stage, axis=1)
pokedata.tail()
```

```python
evol = pokedata.groupby('evolution_stage').count()['id']
evol
```

```python
fig = make_subplots(rows = 1, cols = 2, specs=[[{'type':'domain'}, {'type':'domain'}]])

fig.add_trace(
    go.Pie(
        labels=typegroup.index, 
        values=typegroup.values,
        marker=dict(colors=pokecolors),
        title="Gen 1 Pokemon by Type"
        ),
        1, 1)

fig.add_trace(
    go.Pie(
        labels=evol.index, 
        values=evol.values,
        title="Gen 1 Pokemon by Evolution Stage"
        ),
        1, 2)

fig.show()
```

```python
pokedata['weight_kg'] = pokedata['weight'].str.replace(' kg', '', regex=False)
pokedata['weight_kg'] = pd.to_numeric(pokedata['weight_kg'])
pokedata.sort_values('weight', inplace = True)
```

```python
pokedata['height_m'] = pokedata['height'].str.replace(' m', '', regex=False)
pokedata['height_m'] = pd.to_numeric(pokedata['height_m'])
pokedata.sort_values('height', inplace = True)
pokedata.head()
```

```python
trace1 = go.Scatter(x=pokedata[pokedata['evolution_stage'] == 'Lone']['height_m'],
                    y=pokedata[pokedata['evolution_stage'] == 'Lone']['weight_kg'],
                    mode='markers',
                    zorder=1,
                    name='Lone',
                    marker=dict(size=10, opacity=.35))

trace2 = go.Scatter(x=pokedata[pokedata['evolution_stage'] == 'Starter']['height_m'],
                    y=pokedata[pokedata['evolution_stage'] == 'Starter']['weight_kg'],
                    mode='markers',
                    zorder=2,
                    name='Starter',
                    marker=dict(size=10, opacity=.35))

trace3 = go.Scatter(x=pokedata[pokedata['evolution_stage'] == 'Middle']['height_m'],
                    y=pokedata[pokedata['evolution_stage'] == 'Middle']['weight_kg'],
                    mode='markers',
                    zorder=3,
                    name='Middle',
                    marker=dict(size=10, opacity=.35))

trace4 = go.Scatter(x=pokedata[pokedata['evolution_stage'] == 'Final']['height_m'],
                    y=pokedata[pokedata['evolution_stage'] == 'Final']['weight_kg'],
                    mode='markers',
                    zorder=4,
                    name='Final',
                    marker=dict(size=10, opacity=.35))


layout = go.Layout(title=dict(text='Pokemon weight and height by Evolution Stage'),
    xaxis=dict(title='Weight (in kg)'),
    yaxis=dict(title='Height (in m)'),
    # hovertemplate="%{pokedata['name']} weight:%{_x}kg, height:%{_y}m"
    )

fig.update_traces(
    text="%{pokedata['name']} weight:%{_x}kg, height:%{_y}m", 
    selector=dict(type='scatter')
)

fig = go.Figure(data=[trace1, trace2, trace3, trace4], layout=layout)

fig.show()
#i want to show which pokemon is what on Hover
```
